```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_csv('final_data_revised.csv')

df
```

```python

# Set up the plot
tickers = df.columns[1:]  # Exclude the 'Date' column
num_tickers = len(tickers)
num_columns = 5  # Number of columns in the plot grid
num_rows = (num_tickers + num_columns - 1) // num_columns  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, 2 * num_rows), constrained_layout=True)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each ticker's data
for i, ticker in enumerate(tickers):
    ax = axes[i]
    ax.plot(df["Date"], df[ticker], marker='o')
    ax.set_title(ticker)
    ax.set_xlabel('Date')
    ax.set_ylabel('Dividends')
    ax.tick_params(axis='x', rotation=45)

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```
